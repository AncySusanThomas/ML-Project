## 1.Time Series Analysis-Sales trends over time to identify seasonal patterns and changes in demand

```python
print(df.columns)
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%d/%m/%y')
```

```python
# Step 1: Remove commas (if any) and convert the column to float
df['Unit Price'] = df['Unit Price'].replace({',': ''}, regex=True).astype(float)

# Step 2: Convert the 'Unit Price' to integer (if you want the integer value)
df['Unit Price'] = df['Unit Price'].astype(int)

# If you want to create a 'Total' column based on 'Unit Price' and 'Order Quantity', 
# you can do it like this:
df['Total'] = df['Unit Price'] * df['Order Quantity']

print(df[['Unit Price', 'Total']])
```

```python
sales_trend = pd.DataFrame(df.groupby('OrderYear')['Total'].sum()).reset_index()
```

```python
sales_trend
```

```python
plt.figure(figsize=(14,8)) 
plt.plot(sales_trend['OrderYear'], sales_trend['Total'], marker='o', linestyle='-', color='b')
plt.title('Sales Trends Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.subplots_adjust(left=0.1, right=0.95, top=0.9, bottom=0.3)
plt.show()
```

**From the above graph we have found that if there is an upward slope, it indicates increasing sales over the years,if the trend is flat, sales have remained consistent,A downward slope would indicate a decline in sales.**

## 2.Sales Channel Comparison - Compare sales performance across different channels

```python
print(df.columns)
```

```python
channel_performance = df.groupby('Sales Channel')['Total'].sum().reset_index()
```

```python
channel_performance
```

```python
plt.figure(figsize=(12,8))
plt.bar(channel_performance['Sales Channel'], channel_performance['Total'], color=['blue', 'green', 'orange'])
plt.title('Sales Performance Across Channels', fontsize=16)
plt.xlabel('Sales Channel', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

**From the above graph Sales Channels contribute the most to total sales, highlighting the dominant and underperforming channels. Dominant channels suggest where to focus resources, while weaker channels may require strategy adjustments to boost performance**

## 3.Product Analysis - Visualise the distribution of sales across different products 

```python
print(df.columns)
```

```python
product_sales = df.groupby('_ProductID')['Total'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Sales Distribution Across Products')
plt.axis('equal')  
plt.show()
```

**From the above plot how sales are distributed across different products. It highlights which products contribute the most to total sales, and which ones have a smaller share.This visualization helps identify your top-performing products and shows the relative importance of each product in terms of sales.**

## 4.Discount Analysis - Analyse the impact of discounts on sales

```python
print(df.columns)
```

```python
discount_analysis = df.groupby('Discount Applied')['Total'].mean().reset_index()
```

```python
discount_analysis
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.scatterplot(data=discount_analysis, x='Discount Applied', y='Total', alpha=0.7, color='blue')
plt.title('Impact of Discounts on Sales', fontsize=16)
plt.xlabel('Discount Applied (%)', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.show()
```

**From the above plot we can understand that the relationship between the percentage of discounts applied and the average total sales. It shows whether higher discounts lead to an increase or decrease in sales, helping identify the optimal discount range for maximizing revenue.**

## 5. Regional Performance - Sales Performance Across Different Regions

```python
print(df.columns)
```

```python
region_sales = df.groupby('WarehouseCode')['Total'].sum().reset_index()
```

```python
region_sales
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.scatterplot(data=region_sales, x='WarehouseCode', y='Total', alpha=0.7, color='blue')
plt.title('Regional Performance ', fontsize=16)
plt.xlabel('WarehouseCode', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.show()
```

**From the above plot we can understand that the sales performance across different regions (represented by WarehouseCode). Each point shows the total sales for a specific warehouse, allowing for easy identification of regions with higher or lower sales. This visualization helps in assessing regional performance and can guide decisions on resource allocation, targeted marketing, or operational improvements in underperforming areas.**
